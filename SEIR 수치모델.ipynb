```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# ODE
def SEIR_model(y, t, beta, sigma, gamma):
    S, E, I, R = y
    dS_dt = -beta*S*I
    dE_dt = beta*S*I -sigma*E
    dI_dt = sigma*E - gamma*I
    dR_dt = gamma*I
    return([dS_dt, dE_dt, dI_dt, dR_dt])
```

```python
#parameter
S0 = 1
E0 = 2/10874
I0 = 28/10874
R0 = 0
beta = 0.5
sigma = 1/14
gamma = 1/14

t = np.linspace(0, 100, 100)
```

```python
# solve
sol = odeint(SEIR_model, [S0, E0, I0, R0], t, args=(beta, sigma, gamma))
sol = np.array(sol)
```

```python
plt.figure()
plt.plot(sol[:, 0], 'g-', label='Susceptibles')
plt.plot(sol[:, 1], 'm-', label='Exposed')
plt.plot(sol[:, 2], 'r-', label='Infectious')
plt.plot(sol[:, 3], 'c-', label='Recovereds')
plt.legend()
plt.xlabel('day')
plt.ylabel('ratio')
plt.grid()
plt.show()
```
